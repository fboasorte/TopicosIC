```python
#iris dataset: Conjunto de dados consistem em 3 tipos diferentes de 
#pétalas de íris (Setosa, Versicolour e Virginica) 
#e comprimento da sépala, armazenados em um numpy 150x4.
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix

iris = datasets.load_iris()
X = iris.data[:, 2:4]
y = iris.target #classificacao
#0 Comprimento da sépala; 1 Largura da sépala; 
#2 comprimento da pétala; 3 Largura da pétala 
#setosa, versicolor, virginica
plt.subplots()
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1)
plt.xlabel('Comprimento petala')
plt.ylabel('Largura petala')
plt.grid(True)
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier

X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.20,random_state=42)


knn = KNeighborsClassifier(n_neighbors = 3).fit(X_train, y_train)

predictknn = knn.predict(X_test)

print(confusion_matrix(y_test, predictknn))  
print(classification_report(y_test, predictknn))
```
